## Looking Back: My CSSE 1 Journey

When I think back on CSSE 1, it’s crazy how much I’ve learned. At the start, I didn’t struggle a lot, but I was still figuring things out and getting used to how everything worked. Now, I can look at what I’ve made and actually feel proud of it.

---

## Comparing Then and Now

At the beginning of the year, I was just starting to really get into computer science. My coding skills weren’t bad, but I didn’t have much experience yet. Now, I feel way more confident when I write or debug code.

**Confidence:** I don’t second-guess myself as much. When I see an error, I know I can figure it out.

**Game Development:** I learned how to build real game mechanics, animations, and player movement. Getting the staircases and scrolling to work was tough, but it was super satisfying when it finally did.

**Problem Solving:** I got a lot better at breaking big problems into smaller ones. Instead of getting stuck, I just take it step by step until it works.

Overall, I went from just learning the basics to actually creating something that people can play and enjoy.

---

## Sprint Reflections

### Tools Sprint: Getting Set Up

This sprint was all about setting up everything we needed to start coding. I learned how to install and configure VSCode, Python, and Ruby. It took a little while to get everything working right, but once I did, it made things way smoother later on.

**Lesson Learned:** Taking time to set things up properly in the beginning saves a lot of time later when you’re working on bigger projects.

---

### Fundamentals of JavaScript Sprint

This sprint was one of my favorites. I learned the main parts of JavaScript like variables, loops, functions, and conditionals. It actually clicked pretty quickly for me, and I had fun experimenting with the code.

**What I Liked:** Seeing how JavaScript can make web pages interactive was awesome. It made me realize how much you can do with just a few lines of code.

---

### Quest of Spook Sprint: The Mansion Game

This was my favorite sprint. I built the whole mansion game myself, and it turned out really cool. I added movement, animation, and different levels to explore.

**Challenges:** Getting the staircases to scroll smoothly took a lot of testing and trial and error. Once I figured it out, it made the game feel way more complete.

**What I Learned:** Building the game alone taught me how to plan things out, stay organized, and solve problems without relying on others.

---

## Night at the Museum

Night at the Museum was awesome. After weeks of working on the game, I finally got to show it to parents, students, and teachers.

**The Presentation:** It felt really good showing something I made completely on my own. Watching people actually play it was super rewarding.

**Feedback:** Everyone liked it a lot. Parents thought it looked really polished and fun to play. My teammate’s mom even took pictures because she was surprised by what we made.

It was cool seeing people enjoy something I built from scratch.

The Saturday before Night at the Museum was intense. I still had things that weren’t fully working, and I knew I had to get it done.

After a lot of hours of debugging and testing, everything finally came together. The staircases worked, the movement was smooth, and the game ran perfectly. It was such a great feeling to see it all finally working.
